## Example usage for flowplotlib

```python
from elements import StartStop, Process, Decision
from flowchart import Flowchart
```

```python
# Create a flowchart grid
flow = Flowchart(rows=4, columns=4, cell_size=0.3, text_size=8)

# Add elements based on grid positions
process1 = flow.add_element(Process("Process 1"), 2, 2)

start = flow.add_element(StartStop("Start"), 1, 2)
process2 = flow.add_element(Process("Process 2"), 2, 1)
process3 = flow.add_element(Process("Process 3"), 2, 3)
decision = flow.add_element(Decision("Decision"), 3, 2)
stop = flow.add_element(StartStop("Stop"), 4, 2)

# Connect elements
flow.connect(start, process1, "S1>P1")
flow.connect(process1, process2, "P1>P2")
flow.connect(process1, process3, "P1>P3")
flow.connect(process1, decision, "P1>D")

flow.connect(start, process2, "S>P2")
flow.connect(process2, decision, "P2>D")
flow.connect(process3, decision, "P3>D")
flow.connect(start, process3, "S>P3")

flow.connect(decision, stop, "D>St")

# Draw the flowchart and save it as SVG and PNG files
flow.draw("example_flowchart")
```
