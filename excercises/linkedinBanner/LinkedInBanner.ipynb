```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
import random

# Generate data
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 40)
X, Y = np.meshgrid(x, y) # Convert to grid coordinates

# Randomize parameters for two ripples
origin1 = (random.uniform(0, 10), random.uniform(0, 10)) 
origin2 = (random.uniform(0, 10), random.uniform(0, 10))

magnitude1 = random.uniform(0.5, 1.5)
magnitude2 = random.uniform(0.5, 1.5)

frequency1 = random.uniform(0.5, 1.5)
frequency2 = random.uniform(0.5, 1.5)

# Create ripples
Z1 = magnitude1 * np.sin(frequency1 * np.sqrt((X - origin1[0])**2 + (Y - origin1[1])**2))
Z2 = magnitude2 * np.sin(frequency2 * np.sqrt((X - origin2[0])**2 + (Y - origin2[1])**2))
Z = Z1 + Z2 # Combine arrays

# Contour plot
plt.figure(figsize=(20, 5))
cm = plt.get_cmap('magma')

contour_levels = np.linspace(Z.min(), Z.max(), 8)
contours = plt.contour(X, Y, Z, levels=contour_levels, colors='black')

plt.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()], origin='lower',
           cmap=cm, aspect='auto')

plt.axis('off')
plt.savefig('banner.png', dpi=300, bbox_inches='tight', pad_inches=0)

plt.show()
```
